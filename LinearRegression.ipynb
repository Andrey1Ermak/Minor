$\hat{y}_i = a_0 + a_1 x_{i1} + a_2 x_{i2} + \dots + a_n x_{in} = a^T x_i = X \mathbf{a}$

$\epsilon_i = y_i - \hat{y}_i = \mathbf{y} - X \mathbf{a}$

$E = \sum_{i=1}^m (y_i - \hat{y}_i)^2 = \|\mathbf{y} - X \mathbf{a}\|^2$

$\frac{\partial E}{\partial \mathbf{a}} = -2 X^T (\mathbf{y} - X \mathbf{a}) = 0$

$X^T \mathbf{y} - X^T X \mathbf{a} = 0$

$X^T X \mathbf{a} = X^T \mathbf{y}$

$\mathbf{a} = (X^T X)^{-1} X^T \mathbf{y}$

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([25, 45, 55, 70, 90, 100, 120, 150]).reshape(-1, 1)  
y = np.array([150, 250, 270, 300, 350, 380, 450, 500]) 

X_b = np.c_[np.ones((X.shape[0], 1)), X]

theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

intercept = theta_best[0]
slope = theta_best[1]

print(f'slope: {slope}')
print(f'intercept: {intercept}')

y_pred = X_b.dot(theta_best)

mse = np.mean((y_pred - y) ** 2)
print(f'mse: {mse}')

plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.show()
```

1. Проблемы с мультиколлинеарностью: 
- для $X^T X$ - X становится вырожденной (или почти вырожденной) матрицей, что делает её обратную матрицу трудно вычислимой.
- Вырожденная матрица — это матрица, у которой определитель равен нулю. 
- Определитель матрицы — это числовая характеристика, которая существует только для квадратных матриц (матриц с одинаковым количеством строк и столбцов). Определитель матрицы — это число, которое вычисляется на основе элементов матрицы и определяет её свойства. 
2. Прямой метод через обратную матрицу не подходит для очень больших наборов данных, так как время работы растет кубически по количеству признаков. Например, если у вас несколько миллионов признаков, этот подход станет слишком дорогим по вычислениям и памяти. Когда количество данных гораздо больше, чем количество признаков, этот метод всё ещё работает, но для ситуаций, когда становятся очень большими, нормальное уравнение не является наилучшим выбором
