```python
import numpy as np
import scipy.cluster.hierarchy as sch
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

np.random.seed(42)
data = np.random.rand(10, 5)

# 2. Вычисление расстояний между объектами
distance_matrix = pdist(data, metric='cityblock') 

# 3. Построение дендрограммы с кастомизацией метода
# Можно выбрать другие методы, например 'single', 'complete', 'average'
linkage_matrix = sch.linkage(distance_matrix, method='complete')  # Метод 'complete' (наибольшая связь)

plt.figure(figsize=(10, 6))
sch.dendrogram(linkage_matrix)
plt.title('Dendrogram (Complete Linkage)')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# 4. Построение тепловой карты с результатами кластеризации
dendro = sch.dendrogram(linkage_matrix, no_plot=True)
leaf_order = dendro['leaves']

# Перемещаем данные согласно порядку кластеров
ordered_data = data[leaf_order, :]

# Создаем тепловую карту
plt.figure(figsize=(10, 6))
sns.heatmap(ordered_data, cmap='coolwarm', annot=True, yticklabels=[f'{i}' for i in leaf_order])
plt.title('Heatmap of Hierarchical Clustering (Complete Linkage)')
plt.show()
```
